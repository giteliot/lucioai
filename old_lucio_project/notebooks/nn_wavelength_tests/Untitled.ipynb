```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
cmd = "VAI"
df = pd.read_csv(f'./{cmd}.csv', header=None, delimiter=";")
df.columns=['wave']
df['cmd']=cmd
df['wave'] = df['wave'].apply(json.loads)
df.head()
```

```python
plt.plot(df['wave'].tolist()[0])
```

```python
plt.plot(df['wave'].tolist()[1])
```

```python
plt.plot(df['wave'].tolist()[2])
```
