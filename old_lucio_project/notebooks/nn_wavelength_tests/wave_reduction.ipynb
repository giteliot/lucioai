```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
cmd = "SALTA"
df = pd.read_csv(f'./{cmd}.csv', header=None, delimiter=";")
df.columns=['wave']
df['cmd']=cmd
df['wave'] = df['wave'].apply(json.loads)
df.head()
```

```python
for w in df['wave'].tolist()[:3]:
    plt.plot(w)
```
